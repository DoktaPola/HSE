```python
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd

dist_prev = [60, 277, 900, 521, 707, 525, 946, 755, 916, 777, 626, 465, 515, 980, 323, 603, 759, 276, 199, 470, 953, 460, 557, 76, 5, 963, 200, 221, 524, 475, 79, 162, 441, 305, 550, 622, 697, 112, 797, 333, 593, 891, 912, 238, 905, 581, 773, 202, 869, 508, 933, 62, 519, 627, 620, 883, 713, 594, 781, 388, 972, 393, 44, 673, 68, 590, 468, 138, 769, 291, 41, 988, 664, 446, 727, 758, 120, 826, 854, 378, 638, 413, 845, 242, 36, 38, 683, 329, 667, 851, 706, 943, 314, 352, 927, 474, 577, 736, 386, 500, 338, 569, 908, 956, 534, 122, 601, 848, 190, 40, 402, 587, 391, 347, 92, 669, 592, 613, 732, 149, 360, 692, 499, 197, 936, 260, 882, 822, 795, 180, 70, 320, 53, 377, 143, 173, 728, 985, 37, 794, 910, 154, 346, 997, 834, 899, 798, 612, 886, 762, 619, 107, 827, 479, 554, 325, 662, 965, 646, 381, 733, 19, 901, 792, 884, 476, 990, 787, 78, 866, 665, 331, 172, 110, 24, 705, 549, 428, 461, 278, 96, 961, 928, 538, 551, 540, 20, 15, 344, 655, 837, 217, 668, 645, 145, 366, 198, 88, 870, 126, 906, 738, 944, 932, 831, 680, 105, 575, 676, 174, 438, 452, 271, 131, 406, 887, 610, 332, 948, 298, 890, 527, 786, 520, 756, 9, 369, 266, 265, 426, 642, 395, 585, 584, 89, 746, 560, 533, 140, 336, 973, 959, 27, 926, 861, 542, 365, 947, 54, 196, 858, 430, 880, 447, 408, 750, 83, 876, 111, 660, 17, 530, 299, 416, 671, 392, 137, 885, 213, 286, 528, 780, 322, 496, 825, 789, 375, 617, 875, 159, 754, 699, 952, 368, 16, 558, 74, 898, 832, 556, 169, 561, 290, 31, 652, 836, 712, 49, 513, 214, 636, 409, 324, 195, 236, 50, 808, 127, 330, 600, 616, 133, 300, 309, 164, 226, 922, 807, 481, 257, 403, 262, 267, 618, 472, 97, 731, 72, 230, 237, 729, 417, 443, 767, 387, 370, 748, 281, 358, 43, 264, 75, 868, 373, 966, 47, 220, 983, 842, 597, 353, 90, 546, 501, 545, 58, 21, 737, 815, 51, 427, 531, 722, 942, 823, 544, 95, 960, 480, 752, 651, 207, 453, 462, 69, 436, 407, 495, 522, 647, 385, 839, 871, 791, 566, 715, 485, 711, 118, 440, 802, 661, 148, 981, 26, 918, 596, 337, 820, 555, 867, 450, 771, 921, 297, 951, 454, 835, 494, 315, 94, 102, 516, 155, 958, 734, 59, 326, 937, 583, 744, 342, 964, 296, 563, 962, 57, 175, 156, 1, 63, 466, 878, 471, 635, 903, 184, 650, 824, 801, 457, 374, 996, 273, 949, 941, 637, 929, 694, 971, 743, 7, 491, 35, 511, 113, 153, 134, 492, 45, 294, 248, 81, 146, 628, 934, 599, 840, 42, 509, 167, 335, 384, 255, 548, 914, 813, 503, 303, 194, 595, 695, 830, 168, 939, 414, 103, 420, 690, 547, 225, 677, 778, 30, 572, 176, 117, 376, 571, 189, 682, 345, 135, 371, 285, 493, 779, 275, 219, 523, 553, 742, 11, 611, 688, 518, 444, 295, 632, 482, 100, 433, 274, 142, 517, 678, 263, 253, 995, 852, 657, 654, 850, 170, 986, 864, 656, 187, 672, 382, 821, 312, 458, 429, 532, 423, 341, 923, 431, 751, 308, 670, 418, 13, 307, 192, 675, 223, 629, 811, 147, 710, 790, 165, 658, 254, 487, 498, 598, 186, 359, 775, 641, 841, 643, 666, 562, 631, 896, 279, 287, 586, 853, 33, 160, 321, 477, 507, 529, 56, 894, 589, 674, 691, 163, 630, 415, 32, 98, 969, 228, 316, 816, 800, 65, 317, 709, 84, 615, 740, 698, 812, 25, 3, 434, 749, 909, 232, 218, 77, 349, 768, 398, 399, 693, 506, 552, 818, 252, 451, 115, 327, 404, 925, 411, 907, 301, 445, 817, 809, 968, 950, 847, 570, 23, 211, 85, 361, 128, 783, 177, 579, 588, 974, 185, 106, 829, 2, 91, 614, 833, 193, 182, 904, 984, 209, 514, 422, 573, 235, 467, 989, 421, 489, 123, 171, 735, 28, 578, 130, 772, 753, 863, 67, 541, 215, 915, 747, 249, 394, 681, 302, 205, 66, 686, 478, 71, 250, 328, 760, 284, 716, 259, 702, 502, 12, 437, 739, 543, 18, 774, 992, 799, 269, 920, 424, 231, 124, 224, 243, 576, 604, 484, 889, 724, 48, 383, 313, 310, 784, 644, 725, 99, 166, 745, 289, 29, 719, 261, 976, 982, 178, 245, 64, 14, 917, 318, 379, 913, 999, 994, 704, 819, 108, 364, 448, 380, 132, 838, 879, 10, 687, 208, 895, 144, 720, 741, 389, 602, 862, 86, 201, 0, 473, 857, 256, 938, 975, 568, 158, 283, 856, 535, 648, 234, 924, 621, 893, 293, 970, 723, 114, 150, 567, 340, 8, 464, 463, 806, 855, 6, 348, 721, 400, 609, 425, 455, 357, 351, 141, 624, 930, 52, 439, 804, 874, 157, 911, 483, 239, 119, 280, 559, 355, 591, 653, 776, 859, 311, 978, 888, 410, 606, 565, 764, 152, 935, 796, 61, 469, 151, 306, 717, 282, 940, 967, 356, 210, 179, 191, 486, 814, 229, 161, 955, 977, 860, 272, 82, 240, 640, 954, 246, 844, 456, 101, 459, 343, 334, 203, 87, 339, 390, 396, 412, 718, 987, 582, 634, 188, 902, 843, 39, 766, 397, 204, 270, 505, 181, 222, 497, 714, 605, 919, 367, 607, 93, 432, 574, 892, 206, 877, 580, 633, 770, 765, 998, 241, 803, 233, 22, 136, 504, 873, 258, 401, 435, 957, 244, 488, 539, 872, 659, 761, 4, 763, 216, 679, 685, 34, 442, 449, 139, 227, 828, 703, 810, 354, 319, 689, 785, 782, 116, 537, 663, 251, 247, 526, 993, 931, 623, 897, 46, 564, 849, 129, 362, 805, 793, 80, 608, 512, 372, 708, 419, 490, 649, 363, 405, 268, 510, 104, 183, 292, 945, 288, 730, 350, 304, 846, 757, 991, 625, 726, 684, 212, 73, 881, 700, 55, 109, 536, 788, 696, 701, 979, 121, 125, 639, 865]

route = [359, 775, 641, 841, 643, 666, 562, 631, 896, 279, 287, 586,
          853, 33, 160, 321, 477, 507, 529, 56, 894, 589, 674, 691, 163, 630, 415, 32,
          98, 969, 228, 316, 816, 800, 65, 317, 709, 84, 615, 740, 698, 812, 25, 3, 434,
          749, 909, 232, 218, 77, 349, 768, 398, 399, 693, 506, 552, 818, 252, 451, 115, 327,
          404, 925, 411, 907, 301, 445, 817, 809, 968, 950, 847, 570, 23, 211, 85, 361, 128,
          783, 177, 579, 588, 974, 185, 106, 829, 2, 91, 614, 833, 193, 182, 904, 984, 209, 514,
          422, 573, 235, 467, 989, 421, 489, 123, 171, 735, 28, 578, 130, 772, 753, 863, 67, 147,
          215, 915, 747, 249, 394, 681, 302, 205, 66, 686, 478, 71, 250, 328, 760, 284, 716, 259,
          702, 502, 12, 437, 739, 543, 18, 774, 992, 799, 269, 920, 424, 231, 124, 224, 243, 576,
          604, 484, 889, 724, 48, 383, 313, 310, 784, 60, 277, 900, 521, 707, 525, 946, 755, 916,
          777, 626, 465, 515, 980, 323, 603, 759, 276, 199, 470, 953, 460, 557, 76, 5, 963, 200, 221, 524,
          475, 79, 162, 441, 305, 550, 622, 697, 112, 797, 333, 593, 891, 912, 238, 905, 581, 773, 202, 869, 508,
          933, 62, 519, 627, 620, 883, 713, 594, 781, 388, 972, 393, 44, 673, 68, 590, 468, 138, 769, 291,
          41, 988, 664, 446, 727, 758, 120, 826, 854, 378, 638, 413, 845, 242, 36, 38, 683, 329, 667, 851, 706,
          943, 314, 352, 927, 474, 577, 736, 386, 500, 338, 569, 908, 956, 534, 122, 601, 848, 190, 40, 402, 587,
          391, 347, 92, 669, 592, 613, 732, 149, 360, 692, 499, 197, 936, 260, 882, 822, 795, 180, 70, 320, 53,
          377, 143, 173, 728, 985, 37, 794, 910, 154, 346, 997, 834, 899, 798, 612, 886, 762, 619, 107, 827, 479,
          554, 325, 662, 965, 646, 381, 733, 19, 901, 792, 884, 476, 990, 787, 78, 866, 665, 331, 172, 110, 24,
          705, 549, 428, 461, 278, 96, 961, 928, 538, 551, 540, 20, 15, 344, 655, 837, 217, 668, 645, 145, 366,
          198, 88, 870, 126, 906, 738, 944, 932, 831, 680, 105, 575, 676, 174, 438, 452, 271, 131, 406, 887, 610,
          332, 948, 298, 890, 527, 786, 520, 756, 9, 369, 266, 265, 426, 642, 395, 585, 584, 89, 746, 560, 533,
          140, 336, 973, 959, 27, 926, 861, 542, 365, 947, 54, 196, 858, 430, 880, 447, 408, 750, 83, 876, 111,
          660, 17, 530, 299, 416, 671, 392, 137, 885, 213, 286, 528, 780, 322, 496, 825, 789, 375, 617, 875, 159,
          754, 699, 952, 368, 16, 558, 74, 898, 832, 556, 169, 561, 290, 31, 652, 836, 712, 49, 513, 214, 636,
          409, 324, 195, 236, 50, 808, 127, 330, 600, 616, 133, 300, 309, 164, 226, 922, 807, 481, 257, 403,
          262, 267, 618, 472, 97, 731, 72, 230, 237, 729, 417, 443, 767, 387, 370, 748, 281, 358, 43, 264, 75,
          868, 373, 966, 47, 220, 983, 842, 597, 353, 90, 546, 501, 545, 58, 21, 737, 815, 51, 427, 531, 722,
          942, 823, 544, 95, 960, 480, 752, 651, 207, 453, 462, 69, 436, 407, 495, 522, 647, 385, 839, 871,
          791, 566, 715, 485, 711, 118, 440, 802, 661, 148, 981, 26, 918, 596, 337, 820, 555, 867, 450, 771,
          921, 297, 951, 454, 835, 494, 315, 94, 102, 516, 155, 958, 734, 59, 326, 937, 583, 744, 342, 964,
          296, 563, 962, 57, 175, 156, 1, 63, 466, 878, 471, 635, 903, 184, 650, 824, 801, 457, 374, 996, 273,
          949, 941, 637, 929, 694, 971, 743, 7, 491, 35, 511, 113, 153, 134, 492, 45, 294, 248, 81, 146, 628, 934,
          599, 840, 42, 509, 167, 335, 384, 255, 548, 914, 813, 503, 303, 194, 595, 695, 830, 168, 939, 414, 103,
          420, 690, 547, 225, 677, 778, 30, 572, 176, 117, 376, 571, 189, 682, 345, 135, 371, 285, 493, 779, 275,
          219, 523, 553, 742, 11, 611, 688, 518, 444, 295, 632, 482, 100, 433, 274, 142, 517, 678, 263, 253, 995,
          852, 657, 654, 850, 170, 986, 864, 656, 187, 672, 382, 821, 312, 458, 429, 532, 423, 341, 923, 431, 751, 308, 670, 418, 13, 307, 192, 675, 223, 629, 811, 541, 710, 790, 165, 658, 254, 487, 498, 598, 186, 644, 725, 99, 166, 745, 289, 29, 719, 261, 976, 982, 178, 245, 64, 14, 917, 318, 379, 913, 999, 994, 704, 819, 108, 364, 448, 380, 132, 838, 879, 10, 687, 208, 895, 144, 720, 741, 389, 602, 862, 86, 201, 0, 473, 857, 256, 938, 975, 568, 158, 283, 856, 535, 648, 234, 924, 621, 893, 293, 970, 723, 114, 150, 567, 340, 8, 464, 463, 806, 855, 6, 348, 721, 400, 609, 425, 455, 357, 351, 141, 624, 930, 52, 439, 804, 874, 157, 911, 483, 239, 119, 280, 559, 355, 591, 653, 776, 859, 311, 978, 888, 410, 606, 565, 764, 152, 935, 796, 61, 469, 151, 306, 717, 282, 940, 967, 356, 210, 179, 191, 486, 814, 229, 161, 955, 977, 860, 272, 82, 240, 640, 954, 246, 844, 456, 101, 459, 343, 334, 203, 87, 339, 390, 396, 412, 718, 987, 582, 634, 188, 902, 843, 39, 766, 397, 204, 270, 505, 181, 222, 497, 714, 605, 919, 367, 607, 93, 432, 574, 892, 206, 877, 580, 633, 770, 765, 998, 241, 803, 233, 22, 136, 504, 873, 258, 401, 435, 957, 244, 488, 539, 872, 659, 761, 4, 763, 216, 679, 685, 34, 442, 449, 139, 227, 828, 703, 810, 354, 319, 689, 785, 782, 116, 537, 663, 251, 247, 526, 993, 931, 623, 897, 46, 564, 849, 129, 362, 805, 793, 80, 608, 512, 372, 708, 419, 490, 649, 363, 405, 268, 510, 104, 183, 292, 945, 288, 730, 350, 304, 846, 757, 991, 625,
          726, 684, 212, 73, 881, 700, 55, 109, 536, 788, 696, 701, 979, 121, 125, 639, 865]

class Point:
    def __init__(self, ind, x, y):
        self.index = ind
        self.x = x
        self.y = y

    def __repr__(self):
        return str(self.index)

def plotTSP(pathes, points, num_iters=1):

    x = list()
    y = list()
    for i in range(0, len(points)):
        x.append(points[i][0])
        y.append(points[i][1])
        
    fig, ax = plt.subplots(figsize=(30,15))
    plt.plot(x, y, 'co', markersize=12)

    a_scale = float(max(x)) / float(100)

    # Draw the older paths, if provided
    if num_iters > 1:

        for i in range(1, num_iters):

            # Transform the old paths into a list of coordinates
            xi = []
            yi = []
            for j in pathes[1]:
                xi.append(j[0])
                yi.append(j[1])



            plt.arrow(xi[-1], yi[-1], (xi[0] - xi[-1]), (yi[0] - yi[-1]),
                      head_width=a_scale, color='r',
                      length_includes_head=True, ls='dashed',
                      width=0.001 / float(num_iters))
            for i in range(0, len(x) - 1):
                plt.arrow(xi[i], yi[i], (xi[i + 1] - xi[i]), (yi[i + 1] - yi[i]),
                          head_width=a_scale, color='r', length_includes_head=True,
                          ls='dashed', width=0.001 / float(num_iters))

    # Draw the primary path for the TSP problem
    plt.arrow(x[-1], y[-1], (x[0] - x[-1]), (y[0] - y[-1]), head_width=a_scale,
              color='g', length_includes_head=True)
    for i in range(0, len(x) - 1):
        plt.arrow(x[i], y[i], (x[i + 1] - x[i]), (y[i + 1] - y[i]), head_width=a_scale,
                  color='g', length_includes_head=True)

    plt.xlim(0, max(x) * 1.1)
    plt.ylim(0, max(y) * 1.1)
    plt.show()


def main():
    df_cities = pd.read_csv('DATA.csv')

    points = list()
    tour_points = list()
    for i in range(0, df_cities.shape[0]):
        tour_points.append(Point(i, x=df_cities.iloc[i]['X'], y=df_cities.iloc[i]['Y']))
        points.append([df_cities.iloc[i]['X'], df_cities.iloc[i]['Y']])

    init_path = route.copy()
    for el in tour_points:
        num = el.index
        init_path[route.index(num)] = (el.x, el.y)

    init_path2 = dist_prev.copy()
    for el in tour_points:
        num = el.index
        init_path2[dist_prev.index(num)] = [el.x, el.y]

    pathes = [init_path, init_path2]

    plotTSP(pathes, init_path, 2)



if __name__ == '__main__':
    main()


```
